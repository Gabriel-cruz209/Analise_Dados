```python
#Para rodar a aplicao
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuarao visual
sns.set_theme(style="darkgrid")

#Carregar base de dados
#Como o arquivo esta na mesma pasta, basta colocar o nome
df = pd.read_csv('SuperStore.csv')

#Mostra as primeiras 5 linhas formatadas
display(df.head())

df.info()

#Anlise Preo
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Sales', bins=30, kde=True, color='skyblue')
plt.title('Distribuio de Preo')
plt.xlabel('Preo')
plt.ylabel('Contagem')
plt.show()

# Comparao Categoria
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Sales',palette='Set2')
plt.title('Preo por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Preo')
plt.show()

#Categoria por Regiao
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Category', hue='Region', fill=True, palette='crest')
plt.title('Categoria por Regio')
plt.xlabel('Categoria')
plt.show()

#Vendas por Segmento
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Sales', hue='Segment', multiple='stack', bins=25, palette='crest')
plt.title('Distribuio de Vendas por Segmento')
plt.xlabel('Vendas')
plt.show()

#Lucro por Categoria e Regiao
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Profit', hue='Region', palette='Set1')
plt.title('Lucro por Categoria Separado por Regio')
plt.xlabel('Categoria')
plt.ylabel('Lucro')
plt.legend(title='Regio')
plt.show()

#Quantidade Categoria e Segmen
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Category', y='Quantity', hue='Segment', split=True, palette='muted')
plt.title('Distribuio da Quantidade por Categoria e Segmento')
plt.xlabel('Categoria')
plt.ylabel('Quantidade')
plt.show()

plt.figure(figsize=(12, 7))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Region', size='Quantity', sizes=(20, 200), alpha=0.6, palette='tab10')
plt.title('Lucro vs. Vendas, Colorido por Regio e Tamanho pela Quantidade')
plt.xlabel('Vendas ($)')
plt.ylabel('Lucro ($)')
plt.show()
```
